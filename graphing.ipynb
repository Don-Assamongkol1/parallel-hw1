```python
import matplotlib.pyplot as plt 
```

# Parallel Overhead

```python
Ns = [16, 32, 64, 128, 256, 512, 1024]
serial_time = [0.012, 0.074, 0.523, 3.785, 28.662001, 217.977997, 1581.965942]
parallel_overhead_time = [0.381, 0.363, 0.831, 3.797, 25.288, 183.037003, 1422.432007]
ratio = [serial_time[i] / parallel_overhead_time[i] for i in range(len(serial_time))]
```

```python
plt.plot(Ns,
ratio,
marker='x'
)

plt.legend(["Ratio"])
plt.title("Ratio of Serial to Parallel for T=1")
plt.xlabel("Number of Vertices (N)")
plt.ylabel("Ratio")
```

# Parallel Speedup

Serial to Parallel Ratio means if it is below 1 and small, that the serial implementation is faster. 
If it is above 1 and larger, it means the parallel version is faster.

```python
# for a given N, contains the speedups for T
Ts = [2, 4, 8, 16, 32, 64]
N_and_T = [
    (16, [0.466, 0.673, 0.942, 1.807, 1.748, 1.817]),
    (32, [0.62, 0.832, 1.401, 2.524, 4.912, 4.834]),
    (64, [1.048, 1.319, 2.177, 4.199, 8.324, 15.998]),
    (128, [3.086, 2.913, 4.169, 8.051, 15.104, 31.223]),
    (256, [14.878, 9.88, 10.205, 19.646, 33.487999, 62.290001]),
    (512, [97.358002, 53.692001, 37.400002, 53.069, 77.764999, 125.466003]),
    (1024, [714.794006, 368.002014, 206.871994, 258.282013, 305.962006, 379.359009])
] 
```

```python
for (i, (N, T_times)) in enumerate(N_and_T):

    # serial_time[i] represents the serial time to execute on N many vertices, 
    # as we use i to index into the Ns array defined above
    speedups = [serial_time[i] / T_time for T_time in T_times] 

    plt.plot(Ts,
    speedups,
    marker='x'
    )

    plt.title(f"Ratio of Serial to Parallel for N={N}")
    plt.xlabel("Number of Threads (T)")
    plt.ylabel("Ratio")
    plt.show()
```
